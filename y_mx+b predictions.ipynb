```python
#import libraries needed
import numpy as np
import pandas as pd
import os
```

### Clean up DataFrame

```python
df = pd.read_csv("CarData.csv")
```

```python
df.drop(df.columns[0],axis=1,inplace=True)
```

```python
df.drop(["th"],axis=1,inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python
df.corr()
```

### Normalize X values

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop("st",axis=1)
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
x = min_max_scaler.fit_transform(x)
```

```python
y = df['st']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

# Model Development

```python
from keras.layers import Input, Dense, Dropout, LeakyReLU
from keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(units=512, input_dim=6,activation=LeakyReLU(alpha=0.15),kernel_initializer="random_uniform"))
model.add(Dropout(0.2))
model.add(Dense(units=256, activation=LeakyReLU(alpha=0.15), kernel_initializer="random_uniform"))
model.add(Dropout(0.15))
model.add(Dense(units=128, activation=LeakyReLU(alpha=0.15), kernel_initializer="random_uniform"))
model.add(Dropout(0.1))
model.add(Dense(units=64, activation=LeakyReLU(alpha=0.15), kernel_initializer="random_uniform"))
model.add(Dropout(0.05))
model.add(Dense(units=32, activation=LeakyReLU(alpha=0.15), kernel_initializer="random_uniform"))
model.add(Dense(units=16, activation=LeakyReLU(alpha=0.15), kernel_initializer="random_uniform"))
model.add(Dense(1, activation='linear'))
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['mse','mae'])
```

```python
model.fit(x_train, y_train,  epochs=250, batch_size=100) 
```

```python
validation = model.evaluate(x_test,y_test)
```

```python
validation
```

```python
y_hat = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_hat)
```
